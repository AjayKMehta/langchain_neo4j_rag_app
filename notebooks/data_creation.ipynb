```python
import numpy as np
import pandas as pd
pd.set_option("display.max.columns", 500)
%matplotlib inline
```

```python
data = pd.read_csv(RAW_DATA_PATH)
data.rename({"Name":"patient_name"}, axis=1, inplace=True)
data["visit_id"] = data.index
data.drop(["Age", "Medication"], axis=1, inplace=True)
data["Insurance Provider"].replace("Medicare", "Medicaid", inplace=True)
print(data.shape)
data.head()
```

#### Create dim_patient

```python
np.random.seed(42)

dim_patient = data[["patient_name", "Gender"]].drop_duplicates()
dim_patient.columns = ["patient_name", "patient_sex"]
dim_patient["patient_id"] = dim_patient.index
dim_patient["patient_dob"] = np.random.choice(pd.date_range(start="1930-01-01", end="2003-01-01"), size=len(dim_patient))
dim_patient["patient_blood_type"] = np.random.choice(data["Blood Type"].unique(), size=len(dim_patient))
dim_patient
```

```python
data = pd.merge(dim_patient, data, on=["patient_name"])
data.drop(["Blood Type", "Gender"], axis=1, inplace=True)
data.drop_duplicates(["patient_name", "Date of Admission"], inplace=True)
data
```

#### Create dim_physician

```python
np.random.seed(42)
doctors_use = np.random.choice(data["Doctor"].values, size=500)
dim_physician = pd.DataFrame(doctors_use, columns=["physician_name"])
dim_physician["physician_id"] = dim_physician.index
dim_physician["physician_dob"] = np.random.choice(pd.date_range(start="1950-01-01", end="1995-01-01"), size=len(dim_physician))

dim_physician["physician_grad_year"] = dim_physician["physician_dob"] + pd.DateOffset(years=30)
dim_physician["medical_school"] = np.random.choice(US_MEDICAL_SCHOOLS, size=len(dim_physician))
dim_physician["salary"] = np.random.normal(loc=300_000, scale=30_000, size=len(dim_physician))
dim_physician
```

#### Create dim_payer

```python
dim_payers = data[["Insurance Provider"]].drop_duplicates().reset_index(drop=True)
dim_payers["payer_id"] = dim_payers.index
dim_payers.columns = ["payer_name", "payer_id"]
dim_payers
```

#### Create dim_hospitals

```python
np.random.seed(42)
dim_hospitals = pd.DataFrame({
    "hospital_id": np.arange(30),
    "hospital_name": data["Hospital"].unique()[0:30],
    "hospital_state": np.random.choice(["TX", "GA", "FL", "CO", "NC"], size=30)
})
dim_hospitals
```

#### Create visits 

```python
np.random.seed(42)
data_visits = data[["patient_id",
                    "Date of Admission",
                    "Billing Amount", 
                    "Room Number",
                    "Admission Type",
                    "Discharge Date",
                    "Test Results",
                    "visit_id"]]

data_visits["physician_id"] = np.random.choice(dim_physician["physician_id"].values, size=len(data_visits))
data_visits["payer_id"] = np.random.choice(dim_payers["payer_id"].values, size=len(data_visits))
data_visits["hospital_id"] = np.random.choice(dim_hospitals["hospital_id"].values, size=len(data_visits))
def clean_column_name(column_name):
    return column_name.lower().replace(' ', '_')

data_visits.rename(columns=clean_column_name, inplace=True)
data_visits
```

```python
text_df_full = pd.DataFrame(
    {
        "chief_complaint": CHIEF_COMPLAINTS,
        "treatment_description": TREATMENT_DESCRIPTIONS,
        "primary_diagnosis": DIAGNOSES
    }
)
text_df_full = text_df_full.sample(3000, replace=True, random_state=101)
text_df_full
```

```python
null_rows = [np.nan]*(len(data_visits) - len(text_df_full))
text_df_null = pd.DataFrame({"chief_complaint": null_rows,
 "treatment_description": null_rows,
 "primary_diagnosis": null_rows})

text_df = pd.concat([text_df_null, text_df_full])
text_df = text_df.sample(len(text_df), random_state=101)
text_df["visit_id"] = data_visits["visit_id"].values

data_visits = pd.merge(data_visits, text_df, on="visit_id")
data_visits
```

```python
np.random.seed(42)
random_vists = data_visits[pd.to_datetime(data_visits["date_of_admission"]).dt.year == 2023].sample(500, random_state=101).index
data_visits.iloc[random_vists, data_visits.columns.get_loc("discharge_date")] = np.nan
data_visits["visit_status"] = np.where(data_visits["discharge_date"].isna(), "OPEN", "DISCHARGED")
data_visits
```
